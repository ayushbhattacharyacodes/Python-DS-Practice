## Exploratory Data Analysis

### Gender

```python
ax_gender = sns.countplot(data=data,x='Gender')

for bars in ax_gender.containers:
        ax_gender.bar_label(bars)
```

```python
sales_gen = data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax_sales_gen = sns.barplot(x='Gender',y='Amount',data=sales_gen)

for bars in ax_sales_gen.containers:
    ax_sales_gen.bar_label(bars)
```

#### From the above graph we can conclude that females have purchased more than male and females have spent more than male

### Age

```python
plt.figure(figsize=(15,5))
ax_age = sns.countplot(x='Age',data=data)
for bar in ax_age.containers:
    ax_age.bar_label(bars)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,5))

purchase_by_age = data.groupby(['Age'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax_pur_age=sns.barplot(x='Age',y='Amount',data=purchase_by_age)

plt.xticks(rotation=90)
plt.show()
```

#### From the above graphs one can infer that people of age 18-55 have made most of the purchases while people of age 30 have spent the most

### Age Groups

```python
count_age_group = data['Age Group'].value_counts(ascending=False).index

ax_age_group = sns.countplot(x='Age Group',data=data,order=count_age_group)

for bars in ax_age_group.containers:
    ax_age_group.bar_label(bars)
```

```python
ax_age_group_gen = sns.countplot(x='Age Group',data=data,hue='Gender')

for bars in ax_age_group_gen.containers:
    ax_age_group_gen.bar_label(bars)
```

```python
sales_age = data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

#### From the above graphs we can conclude that most of the buyers are at the age-group of 26-35 female

### State

```python
state_dict = dict(data['State'].value_counts())
keys=list(state_dict.keys())
values=list(state_dict.values())

plt.figure(figsize=(10,5))
ax_state = sns.barplot(x=keys,y=values)

for bars in ax_state.containers:
    ax_state.bar_label(bars)

plt.xticks(rotation=90)
plt.show()
```

```python
state_sales = data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = state_sales,x='State',y='Orders')
```

```python
state_sales = data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = state_sales,x='State',y='Amount')
```

#### From the above graphs we can infer that Karnataka Maharashtra and Uttar Pradesh had placed most of the orders

### Marital Status

```python
sns.set(rc={'figure.figsize':(10,5)})
ax_marital = sns.countplot(data=data,x='Marital_Status')
for bars in ax_marital.containers:
    ax_marital.bar_label(bars)
```

```python
sales_by_marital_gender = data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data=sales_by_marital_gender,x='Marital_Status',y='Amount',hue='Gender')
```

#### From the above graphs we can see that most of the buyers are married(women) and they have higher purchasing power 

### Occupation

```python
sns.set(rc={'figure.figsize':(15,8)})

ax_occupation = sns.countplot(data=data,x='Occupation')

for bars in ax_occupation.containers:
    ax_occupation.bar_label(bars)
```

```python
spending_by_occupation = data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=spending_by_occupation)
```

#### From the above graph we can observe that people from the IT, Healthcare and Aviation sector have done most of the purchasing

### Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax_product_category = sns.countplot(data=data,x='Product_Category')

for bars in ax_product_category.containers:
    ax_product_category.bar_label(bars)
```

```python
product_category_sales = data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data=product_category_sales,x='Product_Category',y='Amount')
```

#### From the above graphs we can see that most of the sold products are from Food,Clothing and Apparel and Electronics Category

#### Top 10 most sold products

```python
sales_state = data.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_ID',y='Orders')
```

## Conclusion

Individuals aged between 18-55 have made most of the purchases.

Married women between the age of 26-35 have spent mostly on Diwali

Out of employed individuals, people from IT, Healthcare and Aviation sector have done most of the shopping

People mostly prefer buying Clothes Electronics and Food items during Diwali Sales

Most of the sales came from Maharashtra Karnataka and Uttar Pradesh
